```
# data to plot
n_groups = 4
least_important = (6, 2, 2, 5)
little_bit_of_importance = (51, 17, 9, 21)
important_enough = (65, 95, 74, 81)
highly_important = (53, 61, 90, 68)


# create plot
fig, ax = plt.subplots(figsize=(10,8))
index = np.arange(n_groups)
bar_width = 0.2
opacity = 0.8

rects1 = plt.bar(index, least_important, bar_width,
alpha=opacity,
color='b',
label='Least Important')

rects2 = plt.bar(index + bar_width, little_bit_of_importance, bar_width,
alpha=opacity,
color='g',
label='Little Bit Of Importance')

rects3 = plt.bar(index+ (2*bar_width), important_enough, bar_width, 
alpha=opacity,
color='y',
label='Important Enough')

rects4 = plt.bar(index+(3*bar_width), highly_important, bar_width,
alpha=opacity,
color= 'c',
label='Highly Important')

plt.xlabel('Criteria',fontsize=15)
plt.ylabel('Number of People',fontsize=15)
plt.title('Indicate The Order Of Importance Of The Following Criteria While Choosing An OTT Platform',fontsize=15)
plt.xticks(index + bar_width, ('Price', 'Speed', 'Content', 'Convenience'),rotation=45,fontsize=15)
plt.legend(fontsize=12)

plt.tight_layout()
plt.show()
```
